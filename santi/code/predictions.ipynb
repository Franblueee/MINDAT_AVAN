```python
import os
import numpy as np
import matplotlib.pyplot as plt
import re
import tensorflow as tf
import pandas as pd
from skimage.io import imread
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import save_model, load_model
import seaborn as sns
sns.set_theme()

path_data = 'data/'
path_test='Test 2/'
path_models='gdrive/MineriaIII/models/'
#path_models='drive/MyDrive/MineriaIII/models/'
path_pred='gdrive/MineriaIII/predictions/'
#path_pred='drive/MyDrive/MineriaIII/predictions/'
ext_sub = '.csv'
```

```python
test_data = []
test_ids = []

for f in os.listdir(path_data + path_test):
    try:
        img = imread(path_data + path_test + f)
        test_data.append(img)
        test_ids.append(f)
    except:
        print("Cannot read file: " + f)
        
test_data = np.array(test_data)
test_data_prep = tf.keras.applications.vgg19.preprocess_input(test_data)
```

```python
def predictions_file(model_name, file='predictions'):
  model = load_model(path_models + model_name)
  test_preds = np.argmax(model.predict(test_data_prep), axis=-1) + 1
  output = pd.DataFrame({'id.jpg': test_ids, 'label': test_preds})
  output.to_csv(path_pred + file + '.csv', header=True, index=False)
```

```python
predictions_file('all_1024-512-128_ft_v2.h5', file='all_1024-512-128_ft_v2')
```
